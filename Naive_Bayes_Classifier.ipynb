#Naive Bayes Classifier

##80-20

###Gaussian Naive Bayes Classifier

```python
#dividing data into dependent and independent variable
X=df.drop(['spam'], axis=1).values
y=df['spam'].values

#splitting data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6) 

#defining the model
clf = GaussianNB()  

#training the model
model1 = clf.fit(X_train, y_train) 

#making predictions
y_pred = model1.predict(X_test)

#creating a comparison dataframe

comparison_frame = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

comparison_frame.head(5)
```

```python
#evaluating the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

the model was ble to classify 368 and 315 messages correctly.

it classified 147 and 12 messages wrongly.

the model accuracy is 81%,this is not bad.

###Multinomial Naive Bayes Classifier

```python
#defining the model and training 
model2 = MultinomialNB().fit(X_train, y_train)

#making predictions
y_pred1 = model2.predict(X_test)

#creating a comparison dataframe

comparison_frame = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred1})

comparison_frame.head(5)
```

```python
#evaluating the model

print(confusion_matrix(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```

the accuracy dropped 

also this model classified lesser messages correctly.

###Bernoulli Naive Bayes Classifier

```python
#defining the model and training 
model3 = BernoulliNB().fit(X_train, y_train)

#making predictions
y_pred2 = model3.predict(X_test)

#creating a comparison dataframe

comparison_frame = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred2})

comparison_frame.head(5)
```

```python
#evaluating the model
print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

this model has the highest accuracy among the three.
